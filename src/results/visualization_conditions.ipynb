```python
def filter_results_by_field(results, field, value, extra_filters=None):
    """Generic filter: select results where field==value and (optionally more filters)."""
    filtered = results[results[field] == value]
    if extra_filters is not None:
        for k, v in extra_filters.items():
            filtered = filtered[filtered[k] == v]
    return filtered

def get_result_groups(results, groupby_field, model_noise=0.0, temporal_shift=0.0):
    """
    Returns a dictionary of DataFrames grouped by the groupby_field (e.g. condition, or pretrained_success_rate).
    Only rows that match model_noise and temporal_shift are included.
    """
    unique_values = results[groupby_field].unique()
    result_groups = {}
    for val in unique_values:
        filtered = filter_results_by_field(
            results, 
            groupby_field, 
            val, 
            extra_filters={
                'parameters_mlp_model_noise': model_noise, 
                'parameters_neural_temporal_shift': temporal_shift
            }
        )
        # Optionally: fallback logic for legacy field names, customize if needed
        if filtered.empty and groupby_field == 'parameters_experiment_condition' and val == "Baseline-ER":
            filtered = filter_results_by_field(
                results, groupby_field, "Baseline-PER",
                extra_filters={
                    'parameters_mlp_model_noise': model_noise, 
                    'parameters_neural_temporal_shift': temporal_shift
                }
            )
        result_groups[val] = filtered
    return result_groups

# Choose grouping field: e.g. 'parameters_experiment_condition' or 'parameters_experiment_pretrained_success_rate'
# Just change GROUPBY_FIELD to switch seamlessly.
GROUPBY_FIELD = 'parameters_experiment_pretrained_success_rate'  # or 'parameters_experiment_pretrained_success_rate', etc.

# Optionally set filtering options:
MODEL_NOISE = 0.0
TEMPORAL_SHIFT = 0.0

# Generate dictionary for plotting: keys are unique values of GROUPBY_FIELD, values are filtered DataFrames
condition_results = get_result_groups(results, GROUPBY_FIELD, model_noise=MODEL_NOISE, temporal_shift=TEMPORAL_SHIFT)

# Now plot using condition_results. To switch, just change GROUPBY_FIELD above.
```

```python
import matplotlib.pyplot as plt

ERROR_BAND = "std"  # "sem", "std", or "ci95"

import numpy as np
import ast



def slice_plot_range(arr, start_idx=None, end_idx=None):
    """Slice a results series for plotting; None start/end = full array."""
    arr = np.asarray(arr, dtype=float)
    n = len(arr)
    start = 0 if start_idx is None else int(start_idx)
    end = n if end_idx is None else int(end_idx)
    start = max(0, min(start, n))
    end = max(start, min(end, n))
    return arr[start:end], start, end


def t_critical_975(df: int) -> float:
    """Two-sided 95% t critical value (scalar n_seeds → df = n-1)."""
    if df < 1:
        return np.nan
    if df >= 30:
        return 1.96
    table = {
        1: 12.706, 2: 4.303, 3: 3.182, 4: 2.776, 5: 2.571,
        6: 2.447, 7: 2.365, 8: 2.306, 9: 2.262, 10: 2.228,
        11: 2.201, 12: 2.179, 13: 2.160, 14: 2.145, 15: 2.131,
        20: 2.086, 25: 2.060, 30: 2.042,
    }
    if df in table:
        return table[df]
    lower = max(k for k in table if k <= df)
    return table[lower]


def error_band_halfwidth(std, sem, n_seeds, kind=None):
    """Half-width of shaded error band around the mean curve."""
    kind = (kind or ERROR_BAND).lower()
    std = np.asarray(std, dtype=float)
    sem = np.asarray(sem, dtype=float)
    if kind == "sem":
        return sem
    if kind == "std":
        return std
    if kind == "ci95":
        df = max(int(n_seeds) - 1, 0)
        if df < 1:
            return np.zeros_like(sem)
        return t_critical_975(df) * sem
    raise ValueError(f"ERROR_BAND must be 'sem', 'std', or 'ci95'; got {kind!r}")

def plot_conditions_by_pretrained(results, 
                                  condition_field='parameters_experiment_condition', 
                                  pretrained_field='parameters_experiment_pretrained_success_rate',
                                  metric='results_success_rate',
                                  window_size=10,
                                  model_noise=0.0,
                                  temporal_shift=0.0,
                                  figsize=(12, 5),
                                  start_idx=None,
                                  end_idx=None,
                                  error_kind=None):
    """
    For each unique value of pretrained_success_rate, plot all conditions in one figure.
    Separate plot for each pretrained value. All conditions shown per plot.
    """
    def moving_average(arr, window_size=50):
        arr = np.array(arr, dtype=float)
        if len(arr) < window_size:
            return arr
        ret = np.cumsum(np.nan_to_num(arr), dtype=float)
        ret[window_size:] = ret[window_size:] - ret[:-window_size]
        return ret[window_size - 1:] / window_size

    # Filter results by model_noise and temporal_shift
    mask = True
    if 'parameters_mlp_model_noise' in results.columns:
        mask = mask & (results['parameters_mlp_model_noise'] == model_noise)
    if 'parameters_neural_temporal_shift' in results.columns:
        mask = mask & (results['parameters_neural_temporal_shift'] == temporal_shift)
    results_filt = results[mask]

    pretrained_values = sorted(results_filt[pretrained_field].dropna().unique())

    for pretrained_val in pretrained_values:
        df_pre = results_filt[results_filt[pretrained_field] == pretrained_val]
        condition_values = sorted(df_pre[condition_field].dropna().unique())
        plt.figure(figsize=figsize)
        for cond in condition_values:
            df_cond = df_pre[df_pre[condition_field] == cond]
            if df_cond.empty:
                continue
            all_metric_series = []
            for val in df_cond[metric]:
                if isinstance(val, str):
                    try:
                        arr = ast.literal_eval(val)
                    except Exception:
                        arr = []
                else:
                    arr = val
                all_metric_series.append(np.array(arr, dtype=np.float32))
            if not all_metric_series:
                continue
            # Pad all series to the same length for stacking
            max_len = max(len(arr) for arr in all_metric_series)
            padded = [np.pad(arr, (0, max_len - len(arr)), constant_values=np.nan) for arr in all_metric_series]
            stacked = np.vstack(padded)  # shape: (num_trials, num_episodes)
            # moving-average each trial
            ma_stacked = np.array([moving_average(row, window_size) for row in stacked])
            mean = np.nanmean(ma_stacked, axis=0)
            std = np.nanstd(ma_stacked, axis=0)
            n_trials = ma_stacked.shape[0]
            sem = std / np.sqrt(n_trials) if n_trials > 0 else std
            mean, start, end = slice_plot_range(mean, start_idx, end_idx)
            std, _, _ = slice_plot_range(std, start_idx, end_idx)
            sem, _, _ = slice_plot_range(sem, start_idx, end_idx)
            band = error_band_halfwidth(std, sem, n_trials, kind=error_kind)
            steps = np.arange(start, end)
            plt.plot(steps, mean, label=cond)
            plt.fill_between(steps, mean - band, mean + band, alpha=0.2)
        plt.xlabel('Episode')
        plt.ylabel(metric)
        plt.title(f'{metric} per episode (shaded: {ERROR_BAND_LABELS.get((error_kind or ERROR_BAND).lower(), ERROR_BAND)} across trials)\n(pretrained_success_rate={pretrained_val})')
        plt.legend()
        plt.tight_layout()
        plt.show()

# Usage: separates plots by pretrained value, within each plot: all conditions
plot_conditions_by_pretrained(
    results,
    condition_field='parameters_experiment_condition',
    pretrained_field='parameters_experiment_pretrained_success_rate',
    metric='results_success_rate',
    window_size=1,
    model_noise=MODEL_NOISE,
    temporal_shift=TEMPORAL_SHIFT
)
```
